```python
count_cols = ['2022 Population', '2020 Population', '2015 Population', '2010 Population',
              '2000 Population', '1990 Population', '1980 Population', '1970 Population']

for column in count_cols:
    # Calculate Q1, Q3, and IQR
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    
    # Define the lower and upper bounds for outliers
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # Count the number of values that are outside the bounds (outliers)
    outlier_count = ((df[column] < lower_bound) | (df[column] > upper_bound)).sum()
    
    # Print the number of outliers
    print(f'Number of outliers in {column}:', outlier_count)
```

```python
# Calculate outlier for year 2022 
q1 = df['2022 Population'].quantile(0.25)
q3 = df['2022 Population'].quantile(0.75)
iqr = q3 - q1
    
# Define the lower and upper bounds for outliers
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Check the values that are outside the bounds (outliers)
outlier=df[((df['2022 Population'] < lower_bound) | (df['2022 Population'] > upper_bound))]
```

```python
df_pop = df.groupby('Continent')['2022 Population'].sum().sort_values(ascending=False).reset_index()
df_pop
```

```python
df_Area = df.groupby('Continent')['Area (km²)'].sum().sort_values(ascending=False).reset_index()
df_Area
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

sns.barplot(x='Continent', y='2022 Population', data=df_pop, ax=axes[0])
axes[0].set_title('2022 Population by Continent')  

sns.barplot(x='Continent', y='Area (km²)', data=df_Area, ax=axes[1])
axes[1].set_title('Area by Continent')  

plt.tight_layout()
plt.show()

```
