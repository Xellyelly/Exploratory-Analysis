```python
sns.heatmap(df.isnull())
```

## sorting missing values

```python
df['fuel_type']=df['fuel_type'].replace('–','others' )
df['fuel_type']=df['fuel_type'].fillna('unknown')
df['accident']=df['accident'].fillna('unknown')
df['clean_title']=df['clean_title'].fillna('unkown')
```

## Univariate Analysis 


### 1)Categorical

```python
Top_10_brand = df['brand'].value_counts().head(10)
Top_10_model = df['model'].value_counts().head(10)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))
sns.barplot( Top_10_brand ,ax=axes[0])
axes[0].set_title('Top 10 brand', fontsize=14, fontweight='bold')
axes[0].set_xlabel('brand', fontsize=12)
axes[0].set_ylabel('count', fontsize=12)
axes[0].tick_params(axis='x', rotation=45)

sns.barplot( Top_10_model ,ax=axes[1])
axes[1].set_title('Top 10 model', fontsize=14, fontweight='bold')
axes[1].set_xlabel('model', fontsize=12)
axes[1].set_ylabel('count', fontsize=12)
axes[1].tick_params(axis='x', rotation=45)

```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
sns.barplot(df['accident'].value_counts() ,ax=axes[0])
axes[0].set_title('Accident counts ', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Accident', fontsize=12)
axes[0].set_ylabel('count', fontsize=12)
axes[0].tick_params(axis='x', rotation=45)


sns.barplot( df['fuel_type'].value_counts() ,ax=axes[1])
axes[1].set_title('Clean Title counts', fontsize=14, fontweight='bold')
axes[1].set_xlabel('clean_title', fontsize=12)
axes[1].set_ylabel('count', fontsize=12)
axes[1].tick_params(axis='x', rotation=45)

```
